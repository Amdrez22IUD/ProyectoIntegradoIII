## 2. Carga del Dataset

```python
import pandas as pd
from ydata_profiling import ProfileReport

df = pd.read_csv('/content/diabetes.csv')

df = df.rename(columns={
    'Pregnancies': 'Embarazos',
    'Glucose': 'Glucosa',
    'BloodPressure': 'PresionArterial',
    'SkinThickness': 'GrosorPiel',
    'Insulin': 'Insulina',
    'BMI': 'IndiceMasaCorporal',
    'DiabetesPedigreeFunction': 'FuncionPedigree',
    'Age': 'Edad',
    'Outcome': 'Diagnostico'
})

print(f'Dimensiones del dataset: {df.shape[0]} filas x {df.shape[1]} columnas')
df.head(10)
```

## 3. Exploración Básica

```python
print('=== Tipos de datos ===')
print(df.dtypes)
print()
print('=== Valores nulos por columna ===')
print(df.isnull().sum())
print()
print('=== Registros duplicados ===')
print(f'{df.duplicated().sum()} duplicados encontrados')
```

```python
print('=== Estadísticas descriptivas ===')
df.describe()
```

### Nota sobre valores 0 como nulos implícitos

En este dataset, las columnas `Glucosa`, `PresionArterial`, `GrosorPiel`, `Insulina` e `IndiceMasaCorporal` contienen valores `0` que **no son fisiológicamente posibles**. Estos representan datos faltantes codificados como cero, y deben tratarse como nulos en etapas posteriores de preprocesamiento.

```python
cols_con_nulos_implicitos = ['Glucosa', 'PresionArterial', 'GrosorPiel', 'Insulina', 'IndiceMasaCorporal']
print('=== Conteo de ceros (nulos implícitos) ===')
for col in cols_con_nulos_implicitos:
    zeros = (df[col] == 0).sum()
    pct = zeros / len(df) * 100
    print(f'  {col}: {zeros} ceros ({pct:.1f}%)')
```

## 4. Balance de la Variable Objetivo

```python
print('=== Distribución de Diagnostico ===')
counts = df['Diagnostico'].value_counts()
pcts = df['Diagnostico'].value_counts(normalize=True) * 100
print(f'  No diabético (0): {counts[0]} registros ({pcts[0]:.1f}%)')
print(f'  Diabético    (1): {counts[1]} registros ({pcts[1]:.1f}%)')
```

## 5. Reporte Completo con Pandas Profiling

El siguiente reporte genera de forma automática:
- Distribución de cada variable (histogramas)
- Detección de valores faltantes y outliers
- Matriz de correlaciones
- Interacciones entre variables
- Estadísticas descriptivas completas

```python
profile = ProfileReport(
    df,
    title='Exploración: Pima Indians Diabetes Dataset',
    explorative=True
)
profile.to_notebook_iframe()
```

```python
profile.to_file('diabetes_report.html')
print('Reporte guardado en: diabetes_report.html')
```

## 6. Análisis de Correlaciones con la Variable Objetivo

```python
print('=== Correlación de cada variable con Diagnostico ===')
correlaciones = df.corr()['Diagnostico'].drop('Diagnostico').sort_values(ascending=False)
print(correlaciones.to_string())
```

## 7. Conclusiones y Hallazgos

### 7.1 Calidad del Dataset

El dataset tiene 768 registros y 9 variables, sin valores nulos explícitos ni filas duplicadas. Sin embargo, el reporte de Pandas Profiling detectó **9 alertas** relacionadas con ceros que no tienen sentido clínico:

| Variable | Ceros | Porcentaje |
|---|---|---|
| Insulina | 374 | 48.7% |
| GrosorPiel | 227 | 29.6% |
| Embarazos | 111 | 14.5% |
| PresionArterial | 35 | 4.6% |
| IndiceMasaCorporal | 11 | 1.4% |

La variable con mayor problema es `Insulina`, donde casi la mitad de los registros tienen valor 0. Estos valores deben tratarse como datos faltantes en etapas posteriores.

### 7.2 Correlaciones con el Diagnóstico

Las variables con mayor correlación con `Diagnostico` fueron:

1. **Glucosa (0.47)** — es la variable más relacionada con el diagnóstico de diabetes, lo cual tiene sentido clínico directo.
2. **IndiceMasaCorporal (0.29)** — el sobrepeso es un factor de riesgo conocido.
3. **Edad (0.24)** — a mayor edad, mayor probabilidad de diagnóstico positivo.
4. **Embarazos (0.22)** — guarda relación con la diabetes gestacional.
5. **FuncionPedigree (0.17)** — refleja la predisposición genética familiar.

Las variables `GrosorPiel` (0.07) y `PresionArterial` (0.07) tienen muy baja correlación con el diagnóstico.

### 7.3 Correlaciones entre Variables

El reporte también detectó correlación alta entre pares de variables independientes:
- **Edad ↔ Embarazos**: tiene sentido, ya que a mayor edad hay más probabilidad de haber tenido más embarazos.
- **GrosorPiel ↔ Insulina**: ambas variables miden indicadores metabólicos relacionados entre sí.

Esto puede ser relevante si en el futuro se aplican modelos sensibles a la multicolinealidad.
